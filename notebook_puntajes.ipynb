```python
import pandas as pd
# visualiza todas las columnas
pd.set_option('display.max_columns', None)

df_2020 = pd.read_csv("datasets/ArchivoMatr_Adm2020.csv", sep=";")
# eliminar columnaSs preferencia, pond ayo acad y tipo matricula
df_2020 = df_2020.drop(columns=["CODIGO_UNIV", "PREFERENCIA", "POND_ANYO_ACAD", "TIPO_MATRICULA"])
# Valores de VIA
VIA ={
    1: "Regular",
    2: "BEA",
    3: "PACE"
}
# cambiar valor VIA por diccionario VIA
df_2020["VIA"] = df_2020["VIA"].map(VIA)
# eliminar duplicados
df_2020.duplicated().sum() # no hay duplicados
df_2020.isnull().sum() # no hay nulos

df_codigos = pd.read_excel("datasets/Libro_CódigosADM2020_ArchivoMatricula.xlsx")
df_codigos = df_codigos.drop(columns=["AGNO_ACAD", "UNI_SIGLA", "MAX_POST_UNIV", "EXCLUYE_DESDE_PREFERENCIA", "NOMBRE_LARGO", "REGION_PROV",
                                      "PEDAGOGIA", "VACANTES_1SEM", "SC_1S", "VACANTES_2SEM", "SC_2S", "TOTAL_CUPOS_CONVOCAR", "PONDERADO_MINIMO", "PROM_MINIMO_LENMAT",
                                      "POND_MIN_ANTES_PE_POND", "PRUEBA_ESPECIAL", "TIPO_PE", "PONDERACION_PE", "RESTRINGE_SEXO", "EXCLUYE_EGRESADO_ANTES", "%_NOTAS",
                                      "%_RANK", "%_LENG", "%_MATE", "%_HYCS", "%_CIEN", "HYCS_CIEN_ALTERNATIVA", "CUPOS_SUPERNUM_BEA"])

df_completo = pd.merge(
    df_2020,
    df_codigos[["CODIGO", "CARRERA", "UNIVERSIDAD", "LUGAR_IMPARTE"]],
    how="left",
    left_on="CODIGO",
    right_on="CODIGO"
)

# guardar el df completo en un csv
df_completo.head(5)
df_completo.to_csv("datasets/puntajes_matricula_2020.csv", index=False)

```

```python
df = pd.read_csv("datasets/puntajes_matricula_2020.csv", sep=",", encoding="utf-8")
df.info()

df["PTJE_POND"] = df["PTJE_POND"].str.replace(",", ".", regex=False).str.strip()
df["PTJE_POND"] = df["PTJE_POND"].astype(float)

# Agrupar por carrera y calcular mínimo y promedio
resumen_puntajes = (
    df.groupby(["UNIVERSIDAD", "CARRERA", "LUGAR_IMPARTE"], as_index=False)
      .agg(
          PUNTAJE_MINIMO=("PTJE_POND", "min"),
          PUNTAJE_PROMEDIO=("PTJE_POND", "mean")
      )
)
resumen_puntajes["PUNTAJE_PROMEDIO"] = resumen_puntajes["PUNTAJE_PROMEDIO"].round(2)
# mostrar el dataframe completo
pd.set_option('display.max_rows', None)
resumen_puntajes
```
